```python
def get_file_list_from_dir(datadir):#
    all_files = os.listdir(os.path.abspath(datadir))
    #data_files = list(filter(lambda file: file.endswith('.jpeg'), all_files))
    return all_files
```

```python
file_list = get_file_list_from_dir(final_path)#getting the list of filenames from the final combined dataset
random.shuffle(file_list)
```

```python
from math import floor
def get_training_and_testing_sets(file_list):
    split = 0.7
    split_index = floor(len(file_list) * split)
    training = file_list[:split_index]
    testing = file_list[split_index:]
    return training, testing
```

```python
training_set,testing_set = get_training_and_testing_sets(file_list) #splitting the dataset to test and training
```

```python
test_target = "C:/Users/akhur/Desktop/Capstone References/Data/Datasets/Test data"
train_target = "C:/Users/akhur/Desktop/Capstone References/Data/Datasets/Train data"
```

```python
#storing the two sets into two seperate folders
for i in range(0,252):
    train_img_name = training_set[i]
    tr_path_name = os.path.join(final_path,train_img_name )
    target_path = os.path.join(train_target,train_img_name )
    shutil.copy2(tr_path_name,target_path)
    print("{} images copied".format(i))
for i in range(0,109):
    test_img_name = testing_set[i]
    test_path_name = os.path.join(final_path,test_img_name )
    target_path = os.path.join(test_target,test_img_name )
    shutil.copy2(test_path_name,target_path)
    print("{} images copied".format(i))
    
```
